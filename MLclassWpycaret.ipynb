```python
import pandas as pd
from pycaret.classification import *
```

```python
dataf = pd.read_csv('heart.csv')
```

```python
dataf.describe()
```

```python
dataf.dtypes
```

Train and evaluate model


```python
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'thal']
```

```python
experiment = setup(dataf, target='target', categorical_features=categorical_features)
```

```python
best_model = compare_models()
```

```python
#test model
predict_model(best_model,dataf.tail())
#predict_model(best_model,dataf.drop('target', axis=1).tail())
```

```python
save_model(best_model, model_name='ridge-model')
```
